```python
class L:
    def __init__(self):
        self.size=1
        self.element=0
        self.array=self.__newarray(self.size)
        
    def __newarray(self,size):
        return (size*ctypes.py_object)()  
        
    def __len__(self):
        return self.element
    
    def append(self,item):
        if self.size==self.element:
            self.__resize(self.size*2)
        
        self.array[self.element]=item
        self.element=self.element+1
            
    def __resize(self,newsize):
        _a=self.__newarray(newsize)
        self.size=newsize
        for i in range(self.element):
            _a[i]=self.array[i]
        self.array=_a
    
    def __str__(self):
        _a=''
        for i in range(self.element):
            _a+=str(self.array[i])+','
        return '['+_a[:-1]+']'
    
    def __repr__(self):
        return self.__str__()
    
    def __getitem__(self,index):
        if isinstance(index, slice):
            start, stop, step = index.start, index.stop, index.step
            if start is None and stop is None and (step<0):
                start=len(self)-1
                stop = -1
            if start is None:
                start = 0
            if stop is None:
                stop = len(self)
            if step is None:
                step = 1
            _a=L()
            for i in range(start, stop, step):
                _a.append(self.array[i])
            return print(_a)
        elif -self.element<=index<self.element:
            if index<0:
                index=self.element+index
            return self.array[index]
        else:
            return 'index out of range'
    
    def pop(self):
        if self.element==0:return 'out of index'
        print(self.array[self.element-1])
        self.element=self.element-1
        
    def clear(self):
        self.size=1
        self.element=0
        
    def index(self,item):
        for i in range(self.element):
            if self.array[i]==item:
                return i
        return 'not in array'
    
    def insert(self,position,item):
        if self.element==self.size:
            self.__resize(self.size*2)
        if position>=self.element:
            self.array[self.element]=item
            self.element=self.element+1
            return
        for i in range(self.element,position,-1):
            self.array[i]=self.array[i-1]
        self.array[position]=item
        self.element=self.element+1
        
    def __delitem__(self,position):
        if 0<=position<self.element:
            for i in range(position,self.element-1):
                self.array[i]=self.array[i+1]
            self.element=self.element-1
            
    def remove(self,item):
        position=self.index(item)
        if type(position)==int:
            self.__delitem__(position)
        else:
            return position
        
    def sort(self,reverse=False):
        def insertion_sort(arr, left=0, right=None):
            if right is None:
                right = len(arr) - 1

            for i in range(left + 1, right + 1):
                key = arr[i]
                j = i - 1
                while j >= left and arr[j] > key:
                    arr[j + 1] = arr[j]
                    j -= 1
                arr[j + 1] = key

            return arr


        def merge(arr, left, mid, right):
            len1, len2 = mid - left + 1, right - mid
            left_part, right_part = arr[left:mid + 1], arr[mid + 1:right + 1]

            i = j = 0
            k = left

            while i < len1 and j < len2:
                if left_part[i] <= right_part[j]:
                    arr[k] = left_part[i]
                    i += 1
                else:
                    arr[k] = right_part[j]
                    j += 1
                k += 1

            while i < len1:
                arr[k] = left_part[i]
                i += 1
                k += 1

            while j < len2:
                arr[k] = right_part[j]
                j += 1
                k += 1

            return arr


        def tim_sort(arr):
            min_run = 32
            n = len(arr)

            for start in range(0, n, min_run):
                end = min(start + min_run - 1, n - 1)
                arr = insertion_sort(arr, start, end)

            size = min_run
            while size < n:
                for left in range(0, n, size * 2):
                    mid = min(n - 1, left + size - 1)
                    right = min((left + size * 2 - 1), (n - 1))
                    arr = merge(arr, left, mid, right)
                size *= 2

            return arr
        
        
        _a=tim_sort(self.array[:self.element])
        self.clear()
        for item in _a:
            self.append(item)
        if reverse==True:
            self.reverse()
    
    def __iter__(self):
        self.pos = 0
        return self

    def __next__(self):
        if self.pos < self.element:
            value = self.array[self.pos]
            self.pos += 1
            return value
        else:
            raise StopIteration
            
    def reverse(self):
        reversed_array = self.array[:self.element][::-1]
        self.clear()  # Clearing the current array
        for item in reversed_array:
            self.append(item)
        
    def copy(self):
        return self.array[:self.element]
    
    def count(self,item):
        _a=0
        for i in range(self.element):
            if self.array[i]==item:
                _a+=1
        return _a
    
    def extend(self,items):
        for i in items:
            self.append(i)
```
